```python
import os
import random
import matplotlib.pyplot as plt
```

```python
base_path = "/home/aliosman/Desktop/MyProjects/lemon-quality/data/"
```

```python
sample_img_path = base_path + "artifact/0001_E_H_120_C.jpg"
sample_img = plt.imread(sample_img_path)
plt.imshow(sample_img)
```

```python
def random_img_picker(img_dir_path: str):
    return img_dir_path + random.choice(os.listdir(img_dir_path))
```

```python
classes = os.listdir("/home/aliosman/Desktop/MyProjects/lemon-quality/data/")

fig = plt.figure(figsize=(15, 15))

for i in range(0, 7):
    ax = fig.add_subplot(3, 3, i + 1)
    im = plt.imread(random_img_picker(base_path + classes[i] + "/"))
    ax.imshow(im)
    ax.set_title(classes[i])
```
